# Manipulación de datos con Pandas


Pandas es una herramienta de manipulación y análisis de datos de código abierto rápida, potente, flexible y fácil de usar [<https://pandas.pydata.org/>]. Proporciona mecanismos para trabajar con diferentes formatos de datos como archivos CSV (comma-separated values), archivos de Excel o bases de datos. 

Las dos estructuras de datos principales de Pandas son: **Series** (Matriz unidimensional etiquetada de forma homogénea) y **DataFrame** (Estructura de datos bidimensional con columnas que pueden contener diferentes tipos de datos). Podríamos pensar en las estructuras de datos de Pandas como contenedores flexibles para datos de dimensiones inferiores. Por ejemplo, DataFrame es un contenedor para Series y Series es un contenedor para escalares [<https://pandas.pydata.org/pandas-docs/stable/getting_started/overview.html>].

## Selección y filtros

Para seleccionar más de una columna (es decir, para convertir el DataFrame en otro DataFrame), podemos usar doble corchetes [[ ]]

```python
df2 = df[['Muestra', 'Descripcion Muestra', 'Año', 'Atropello']]
df2.head()
```

Si queremos seleccionar filas, podemos usar dos formas, por índice o por posición.

### Por índice

Ya vimos que un DataFrame y una Series tienen índice. Podemos usarlo para seleccionar filas:

```python
# Seleccionamos la fila cuyo valor del índice sea == 0. En Python, los índices se basan en cero.
# .loc[0, ] nos devuelve todas las columnas (equivalente a .loc[0])

df.loc[0, ]
```

```python
df.loc[0, 'Año']
```

```python
df.loc[0, ['Año', 'Descripcion Muestra']]
```

### Por posición

`iloc` nos permite seleccionar por la posición en la tabla:

```python
df.iloc[0, ]
```

¿Cuál es la diferencia?

Si decidimos *alterar* el índice, el índice y la posición ya no serán iguales.

Por ejemplo, podemos filtrar no sólo por el valor del índice, sino que también por alguna condición en las columnas:

```python
# Si filtramos todas las filas cuyo valor de "Año" sea 2010
# el índice se **preserva**

df[df['Año'] == 2010]
```

```python
a2010 = df[df['Año'] == 2010]

a2010.loc[2], a2010.iloc[2]
```

### Más filtros

Pandas predefine muchos operadores booleanos para sus dataframes y series.

```python
# Para hacer un AND de las condiciones (ej. año 2010 y muestra Nacional), se usa & (ampersand o et)
# **siempre** encerrar cada condición entre paréntesis

df[(df['Año'] == 2010) & (df['Muestra'] == 'Nacional')]
```

```python
# Para hacer un OR se ocupa el símbolo | (pipe o tubería)
df[(df['Lesionados Graves'] < 5) | (df['Muertos'] < 4)]
```

```python
# Desigual (!=), mayor o igual (>=), menor o igual (<=), mayor (>), menor (<)
df[df['Muestra'] != 'Comunal']
```

```python
# También podemos filtrar según valores de una lista
df[df['Descripcion Muestra'].isin(['X Región de Los Lagos', 'XI Región de Aysén'])]
```

```python
# Esta condición excluye los registros de las muestras 'Regional' y 'Comunal'
df[~df['Muestra'].isin(['Regional', 'Comunal'])]
```

```python
# Filtrando según el contenido textual
df[df['Descripcion Muestra'].str.contains('Región del')]
```

## Modificación de Dataframes

Veremos cómo eliminar filas o columnas con datos faltantes con `dropna()`.

*   **axis==0** => filas
*   **axis==1** => columnas

**how=='any'** indica que si la fila o columna tiene al menos 1 dato faltante (NA o nan), se borra la fila/columna completa.

Las operaciones de Pandas tienden a producir nuevos dataframes en lugar de modificar los proporcionados. Muchas operaciones tienen el parámetro `inplace` booleano opcional para aplicar los cambios al dataframe en cuestión en lugar de entregar uno nuevo. Es recomendable usar distintas variables para no sobreescribir los datos que estás usando (siempre que puedas).

```python
df2 = df.dropna(axis=1, how='any')
```

También podemos eliminar registros duplicados con la función `drop_duplicates()`. Por defecto elimina los duplicados excepto la primera ocurrencia (parámetro `keep`). 

```python
#df2.duplicated() #permite identificar las filas duplicadas 
df2 = df2.drop_duplicates()
```

A menudo es útil crear nuevas columnas basadas en las existentes utilizando una función. Las nuevas columnas a menudo se denominan características derivadas:

```python
df2['Total_afectados'] = df2[['Muertos', 'Lesionados Graves', 'Lesionados Menos Graves', 'Lesionados Leves']].sum(axis=1)
#df2['Total_afectados'] = df2['Muertos'] +df2['Lesionados Graves'] + df2['Lesionados Menos Graves'] +df2['Lesionados Leves']
#df2['Total_afectados'] = df2.iloc[:, 9:13].sum(axis=1) #rango de índices de las columnas

df2.head()
```

## Agregaciones

```python
df2.groupby(['Año']).mean()
```

```python
# ¿Qué me está diciendo esta tabla?

df2.groupby(['Año', 'Muestra']).sum()
```

```python
# Agrupamos los datos de las comunas por descripción.
# Luego seleccionamos la columna que creamos arriba "Total_afectados".
afectados = df2[df2.Muestra=='Comunal'].groupby(['Descripcion Muestra']).sum()[['Total_afectados']]
afectados
```

```python
# Ahora ordenamos los valores de forma descendente y seleccionamos el top 5.
afectados.sort_values('Total_afectados', ascending=False).head()
```

# Visualizaciones simples con pandas

Algunos ejemplos:

- Histograma / Boxplot: para observar la distribución de un atributo.
- Gráficos de dispersión: para encontrar correlación entre variables posiblemente relacionadas
- Gráfico de barras: cuando una de las variables es categórica
- Gráfico de línea: cuando una variables está estrictamente relacionada con la otra, que es continua (ej. el tiempo)

```python
# Filtraremos los datos Regionales del Año 2011
df_regional = df[(df.Muestra=='Regional') & (df.Año ==2011)]
```

## Boxplot

```python
df_regional[['Atropello', 'Caída', 'Colisión', 'Choque', 'Volcadura', 'Otros ']].plot(kind='box'); 
#plt.ylim(0, 4000)
```

## Histograma

```python
#Los histogramas sirven para agrupar los datos en clases. 
#Luego se cuenta la cantidad de observaciones que hay en cada una de ellas. Por ejemplo:

df_regional.plot(kind='hist', y='Muertos'); #df_regional['Muertos'].hist();
```

## Gráfico de dispersión

```python
df_regional.plot(x='Atropello', y='Muertos', kind='scatter');
```

## Gráfico de barras

```python
ax = df_regional.plot.barh(x='Descripcion Muestra', y='Muertos', figsize=(8, 6), title='Muertes por accidentes durante el 2011')
ax.set_xlabel("Cantidad");
ax.set_ylabel("Región (Descripcion Muestra)");
```

## Gráfico de Línea

```python
dolar = pd.read_csv("https://www.exchange-rates.org/HistoryExchangeRatesReportDownload.aspx?base_iso_code=USD&iso_code=CLP")
dolar
```

```python
# modificamos el atributo fecha
# formatos de fecha: https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes
dolar['Date'] = pd.to_datetime(dolar['Date'], format='%m/%d/%Y %A')

#Graficamos
dolar.plot(kind='line', x='Date', y='Rate');
```
